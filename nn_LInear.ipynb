```python
import torchvision
from torch.utils.data import DataLoader
import torch
from torch.nn import Linear
from torch import nn

dataset = torchvision.datasets.CIFAR10("./dataset",train=False,transform=torchvision.transforms.ToTensor(),download=True)

dataloader = DataLoader(dataset,batch_size=64)

class Model(nn.Module):
    def __init__(self):
        super(Model,self).__init__()
        self.line1 = Linear(in_features=196608,out_features=10)
        
    def forward(self,input):
        output = self.line1(input)
        return output

model = Model()

for data in dataloader:
    imgs,targets = data
    print(imgs.shape)# torch.Size([64, 3, 32, 32])
#     output = torch.reshape(imgs,(1,1,1,-1))
    output1 = torch.flatten(imgs)
#     print(output.shape)# torch.Size([1, 1, 1, 196608])
    print(output1.shape)# torch.Size([196608])
    output1 = model(output1)
#     output = model(output)
#     print(output.shape)# torch.Size([1, 1, 1, 10])
    print(output1.shape)# torch.Size([10])


        
```
